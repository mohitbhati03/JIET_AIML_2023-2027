```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('Toxic.jpg', cv2.IMREAD_GRAYSCALE)

# Gaussian blur (simulate degradation)
blurred = cv2.GaussianBlur(img, (9, 9), 0)

# Restoration techniques
median = cv2.medianBlur(blurred, 5)
bilateral = cv2.bilateralFilter(blurred, 9, 75, 75)

# Show images
cv2_imshow(img)
cv2_imshow(blurred)
cv2_imshow(median)
cv2_imshow(bilateral)
```
