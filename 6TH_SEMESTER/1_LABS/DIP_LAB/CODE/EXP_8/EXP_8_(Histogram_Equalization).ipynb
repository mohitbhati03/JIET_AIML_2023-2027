```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('Dhurandhar.jpeg', cv2.IMREAD_GRAYSCALE)

# Add salt & pepper noise
noise = np.random.choice([0, 255], size=img.shape, p=[0.98, 0.02])
noisy_img = img.copy()
noisy_img[noise == 255] = 255

# Mean filter (Blur)
mean_filtered = cv2.blur(noisy_img, (3, 3))

# Median filter
median_filtered = cv2.medianBlur(noisy_img, 3)

# Show results
cv2_imshow(img)
cv2_imshow(noisy_img)
cv2_imshow(mean_filtered)
cv2_imshow(median_filtered)
```
