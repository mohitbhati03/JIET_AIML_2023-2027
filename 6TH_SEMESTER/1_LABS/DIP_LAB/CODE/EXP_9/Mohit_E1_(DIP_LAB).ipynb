```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read grayscale image
img = cv2.imread("CAMERAMAN_TIF.png", 0)

# ---------------------------
# PREWITT OPERATOR
# ---------------------------

# Define Prewitt kernels
prewitt_x = np.array([[ -1, 0, 1],
                      [ -1, 0, 1],
                      [ -1, 0, 1]])

prewitt_y = np.array([[ 1,  1,  1],
                      [ 0,  0,  0],
                      [-1, -1, -1]])

# Apply filter
prewitt_x_img = cv2.filter2D(img, -1, prewitt_x)
prewitt_y_img = cv2.filter2D(img, -1, prewitt_y)

prewitt_edge = cv2.add(prewitt_x_img, prewitt_y_img)


# ---------------------------
# SOBEL OPERATOR
# ---------------------------

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

sobel_edge = cv2.magnitude(sobel_x, sobel_y)


# ---------------------------
# Display Results
# ---------------------------

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(prewitt_edge, cmap='gray')
plt.title('Prewitt Edge')

plt.subplot(2,2,3)
plt.imshow(sobel_edge, cmap='gray')
plt.title('Sobel Edge')

plt.tight_layout()
plt.show()
```
