```python
import cv2
from google.colab.patches import cv2_imshow
img1 = cv2.imread("CAMERAMAN_TIF.png",0) #image path (GREY --> 0, COLOR --> 1)
cv2_imshow(img1)
```

```python
#50% IMAGE REDUCED
img50 = cv2.resize(img1, None, fx=0.5, fy = 0.5, interpolation=cv2.INTER_AREA)
img50
```

```python
#image converted into jpg
cv2.imwrite("cameraman_reduced.jpg", img50)
```
