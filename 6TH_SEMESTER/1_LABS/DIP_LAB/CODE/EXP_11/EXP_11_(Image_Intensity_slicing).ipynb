```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('spider_man.jpg', cv2.IMREAD_GRAYSCALE)

# Intensity slicing
output = np.zeros_like(img)

# Highlight range 100–150
output[(img >= 100) & (img <= 150)] = 255

# Show
cv2_imshow(img)
cv2_imshow(output)
```
